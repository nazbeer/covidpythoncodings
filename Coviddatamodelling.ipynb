```python
##Answre - 2

import pandas as pd
df=pd.read_excel('CORONO_INDIA_STATE_DATA.xlsx')
state_stats = df.groupby('STATES')['DEATHS', 'CASES'].sum()
state_cases = df.groupby('STATES')['CRITICAL'].sum()
sorted_states = state_cases.sort_values(ascending=False)
top_five_states = sorted_states.head(5)
print("Top Five States with the Most Critical COVID-19 Cases:")
for state, cases in top_five_states.items():
    print(f"{state}: {cases}")
```

```python
##Answer - 3

import pandas as pd
df=pd.read_excel('CORONO_INDIA_STATE_DATA.xlsx')
state_stats = df.groupby('STATES')['DEATHS', 'CASES'].sum()
sorted_states_cases = state_stats['CASES'].sort_values(ascending=False)
sorted_states_deaths = state_stats['DEATHS'].sort_values(ascending=False)
state_with_most_cases = sorted_states_cases.index[0]
most_cases = sorted_states_cases[0]
state_with_most_deaths = sorted_states_deaths.index[0]
most_deaths = sorted_states_deaths[0]
print("State with the Most Cases:")
print(f"{state_with_most_cases}: {most_cases}")
print("\nState with the Most Deaths:")
print(f"{state_with_most_deaths}: {most_deaths}")
```

```python
## Answer - 4 

import pandas as pd
import matplotlib.pyplot as plt
state_active_cases = df.groupby('STATES')['ACTIVE'].sum()
state_active_cases = state_active_cases.sort_values(ascending=False)
top_states = state_active_cases.head(10)
percentages = (top_states / top_states.sum()) * 100
plt.pie(percentages, labels=top_states.index, autopct='%.1f%%', startangle=90)
plt.title('Active Cases by State')
plt.axis('equal')
plt.show()
```

```python
## Answer - 5

import pandas as pd
import matplotlib.pyplot as plt

df['MONTH'] = df['MONTH']

monthly_sum = df.groupby('MONTH').sum()

fig, axs = plt.subplots(2, 3, figsize=(12, 8))

fig.suptitle('COVID-19 Metrics by Month', fontsize=16)

axs[0, 0].plot(monthly_sum.index, monthly_sum['CASES'])
axs[0, 0].set_title('CASES')
axs[0, 0].set_xlabel('MONTH')
axs[0, 0].set_ylabel('Count')

axs[0, 1].plot(monthly_sum.index, monthly_sum['DEATHS'])
axs[0, 1].set_title('DEATHS')
axs[0, 1].set_xlabel('MONTH')
axs[0, 1].set_ylabel('Count')

axs[0, 2].plot(monthly_sum.index, monthly_sum['RECOVERIES'])
axs[0, 2].set_title('RECOVERIES')
axs[0, 2].set_xlabel('MONTH')
axs[0, 2].set_ylabel('Count')

axs[1, 0].plot(monthly_sum.index, monthly_sum['TEST'])
axs[1, 0].set_title('TEST')
axs[1, 0].set_xlabel('MONTH')
axs[1, 0].set_ylabel('Count')

axs[1, 1].plot(monthly_sum.index, monthly_sum['CRITICAL'])
axs[1, 1].set_title('Critical Cases')
axs[1, 1].set_xlabel('MONTH')
axs[1, 1].set_ylabel('Count')

axs[1, 2].plot(monthly_sum.index, monthly_sum['ACTIVE'])
axs[1, 2].set_title('Active Cases')
axs[1, 2].set_xlabel('Month')
axs[1, 2].set_ylabel('Count')

plt.tight_layout()

plt.show()
```
